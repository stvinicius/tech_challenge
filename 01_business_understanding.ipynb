# Tech Challenge Fiap - Fase 1

## Case NPS Preditivo
Prever satisfação do cliente (NPS) antes da aplicação da pesquisa

# Fase 1 - Business Understanding 

## 1. Contexto de Negócio

Um e-commerce nacional passou a lidar com um volume cada vez maior de pedidos, entregas
e interações com clientes. Esse crescimento trouxe ganhos importantes de escala,
mas também revelou desafios relevantes na experiência do cliente, especialmente
refletidos na alta variabilidade do Net Promoter Score (NPS) entre diferentes perfis de
consumidores. A área de Experiência do Cliente percebeu que, mesmo com
indicadores operacionais aparentemente semelhantes, alguns clientes se tornam
promotores da marca, enquanto outros se tornam detratores. 

Atualmente, o NPS é coletado **apenas após o encerramento da jornada de compra**, o que limita a capacidade da empresa de:
- Antecipar problemas antes que afetem a percepção do cliente
- Priorizar ações corretivas nas operações de logística e atendimento
- Atuar de forma **preventiva**, e não apenas reativa

O NPS mede o quanto um cliente recomendaria a empresa a outras pessoas, em uma escala de 0 a 10:

| Categoria | Faixa | Comportamento |
|-----------|-------|---------------|
| **Promotores** | 9 – 10 | Clientes fiéis, recomendam ativamente a marca |
| **Neutros** | 7 – 8 | Satisfeitos, mas sem engajamento ativo |
| **Detratores** | 0 – 6 | Insatisfeitos, podem prejudicar a reputação da marca |

Nesse contexto, surge a necessidade
de transformar dados operacionais, como informações de pedidos, logística e
atendimento, em insights acionáveis, capazes de orientar decisões estratégicas e
operacionais.
 
### Por que o NPS é estratégico para um e-commerce?

#### Impacto na Recompra
Promotores têm probabilidade significativamente maior de realizar novas compras. Em e-commerce, onde o custo de aquisição de clientes (CAC) é elevado, reter um cliente existente é muito mais rentável do que conquistar um novo. Clientes com NPS alto tendem a ter maior LTV (Lifetime Value) e menor churn, impactando diretamente a receita recorrente da operação.

#### Impacto no Boca a Boca
No ambiente digital, a recomendação (ou a crítica) se amplifica exponencialmente: avaliações em marketplaces, posts em redes sociais e comentários em plataformas de review como Reclame Aqui moldam a percepção de novos compradores. Um único detrator vocal pode influenciar dezenas de decisões de compra. Por outro lado, um promotor engajado funciona como canal de marketing orgânico de custo zero.

#### Impacto no Market Share
Em um mercado de e-commerce altamente competitivo, a experiência do cliente é um dos principais diferenciadores. Empresas com NPS consistentemente alto tendem a crescer mais rápido do que o mercado, pois combinam menor custo de retenção com maior taxa de aquisição orgânica. A queda no NPS é um sinal antecedente de perda de market share, ou seja, antes que os números de receita comecem a cair.



 

## 2. Objetivos de Negócio

### Objetivo 1 - Diagnóstico 

A companhia quer entender quais fatores operacionais estão causando essa insatisfação, quais as variaveis estão vinculadas e que mais influenciam a insatisfação dos clientes. Medido pela identificação das top 3 variáveis com maior correlação com NPS ≤ 6.


### Objetivo 2 - Prevenção 

Redução de Detratores em 10% em até 90 dias com o modelo em produção por meio de ações preventivas acionadas automaticamente para clientes identificados como futuros Detratores antes da pesquisa NPS.


## 3. Objetivos analíticos

Os objetivos analíticos traduzem os objetivos de negócio em problemas técnicos de Data Science. 

### Objetivo Analítico 1 — Análise Diagnóstica

> **Realizar uma análise diagnóstica e exploratória (EDA) para identificar quais variáveis operacionais: logísticas (`delivery_delay_days`, `delivery_attempts`), de atendimento (`customer_service_contacts`, `resolution_time_days`, `complaints_count`) e de pedido (`order_value`, `freight_value`), apresentam maior associação com NPS ≤ 6 (Detrator), utilizando estatísticas descritivas, visualizações e análise de correlação.**

- **Tipo de problema:** Descritivo (por que aconteceu?)
- **Técnicas:** EDA, correlação, segmentação por grupos de NPS

### Objetivo Analítico 2 — Modelo Preditivo

> **Construir um modelo de classificação binária que preveja se um pedido (unidade) resultará em avaliação NPS ≤ 6: Detrator (ação) após a conclusão da jornada de compra e antes da aplicação da pesquisa (período), utilizando apenas variáveis operacionais disponíveis no momento da previsão.**

- **Tipo de problema:** Preditivo (classificação binária)
- **Variável-alvo (Target):** `nps_detrator` → 1 se `nps_score ≤ 6`, 0 caso contrário
- **Unidade de análise:** Pedido (`order_id`)
- **Horizonte:** Após entrega e antes da pesquisa NPS
- **Métricas de avaliação:** Recall ≥ 75% (prioridade: capturar detratores), AUC-ROC ≥ 0,80, F1-Score como métrica de equilíbrio


#### Por que Recall é a métrica prioritária?
Em problemas de satisfação do cliente, um **falso negativo** (não identificar um futuro Detrator) é mais custoso do que um **falso positivo** (acionar preventivamente um cliente que não seria Detrator). O custo de oferecer um cupom a um cliente satisfeito é baixo; o custo de deixar um cliente insatisfeito sem ação é alto, em reputação e receita.

#### Variável excluída do modelo
`csat_internal_score` **não deve ser usada como feature do modelo preditivo.**  
Risco de **leakage**: este score interno de satisfação pode ser calculado com base em informações coletadas simultaneamente ou após o NPS, tornando-se uma variável do futuro que o modelo não teria disponível no momento real da previsão. Seu uso inflacionaria artificialmente a performance do modelo em treino.


## 4. Critérios de Sucesso

Os critérios de sucesso definem como saberemos que o projeto funcionou, tanto do ponto de vista técnico quanto do negócio. São a régua que será usada na fase de Avaliação (Evaluation).

### Critérios técnicos (métricas do modelo)

| Métrica | Meta mínima | Justificativa |
|---------|-------------|---------------|
| **Recall** (Detratores) | ≥ 75% | Prioridade capturar quem vai avaliar mal: falso negativo é mais caro |
| **AUC-ROC** | ≥ 0,80 | Capacidade geral de separar Detratores de Não-Detratores |
| **F1-Score** | ≥ 0,70 | Equilíbrio entre Precision e Recall |
| **Precision** (Detratores) | ≥ 60% | Evitar excesso de falsos positivos que sobrecarregue o CRM |

> **Nota:** Acurácia **não** é a métrica principal. Se a base tiver ~25% de Detratores, um modelo trivial (que nunca prevê Detrator) teria 75% de acurácia — sem nenhum valor real.

### Critérios de negócio (impacto real)

| Critério | Meta | Como medir |
|----------|------|------------|
| Redução de Detratores | Reduzir % de Detratores em ≥ 10% em 90 dias | Comparação antes/depois do modelo em produção |
| Efetividade das ações preventivas | ≥ 30% dos clientes acionados pelo modelo evitam avaliar NPS ≤ 6 | Teste A/B: grupo com ação preventiva vs. grupo controle |
| Tempo de resposta | Score gerado em até 24h após a entrega | Monitoramento operacional do pipeline |


## 5. Definição do Target 

### Qual variável representa a satisfação do cliente?
A variável nps_score representa a satisfação do cliente. Ela é uma nota de 0 a 10 fornecida pelo próprio cliente após a conclusão da jornada de compra, respondendo à pergunta clássica do NPS: "Em uma escala de 0 a 10, o quanto você recomendaria nossa empresa a um amigo ou familiar?"  

### Por que ela foi escolhida? 
Unica variavel do Dataser gerado pelo cliente e não pela empresa. Todas outras variáveis são registros operacionais internos. O NPS é a voz direta do cliente sobre sua experiência, o que o torna o indicador mais legítimo de satisfação.  

É uma variável observável, mensurável e com horizonte temporal claro — os três critérios que o CRISP-DM exige de um bom target. Sabemos exatamente quem respondeu, qual nota deu e quando.  

nps_score se conecta diretamente ao objetivo de negócio: reduzir Detratores. Usar nps_score como base do target garante que o modelo responde exatamente à pergunta que o negócio fez, sem aproximações.  

### Em que momento da jornada essa informação é coletada?
Jornada do cliente: pedido realizado → pagamento → separação → envio → entrega → pesquisa NPS 

O nps_score é coletado no último passo, após o encerramento completo da experiência de compra. Isso tem uma implicação crítica para o modelo: ele só pode usar como features as variáveis disponíveis antes da pesquisa ser respondida.  

Variáveis disponíveis antes do NPS e válidas como features: customer_id, order_id, customer_age, customer_region, customer_tenure_months, order_value, items_quantity, discount_value, payment_installments, delivery_time_days, delivery_delay_days, freight_value, delivery_attempts, customer_service_contacts, resolution_time_days, complaints_count.
Variáveis que requerem atenção especial:

**csat_internal_score**: score interno de satisfação gerado pela operação, provavelmente calculado com base em dados que incluem ou são contemporâneos ao NPS. Deve ser excluído do modelo por risco de leakage, ou seja, usá-lo seria como dar ao modelo acesso ao futuro. 

**repeat_purchase_30d**: indica recompra em até 30 dias após o pedido. Merece investigação, se a recompra ocorre antes da pesquisa NPS ser respondida, pode ser uma feature válida. Se ocorre depois, é leakage. A linha do tempo precisa ser verificada antes do uso.


### Existe algum risco de usar essa variável de forma inadequada?

**Leakage (vazamento de dados)**: O risco mais grave neste dataset. O csat_internal_score é um forte candidato a leakage, se ele for gerado com base em informações que só existem após ou simultaneamente ao NPS, incluí-lo como feature faria o modelo "ver o futuro" durante o treino. A performance seria artificialmente alta no notebook e colapsaria em produção. <br>  
**Definição mutante do target**: O corte de Detrator está definido hoje como nps_score ≤ 6. Se a empresa decidir mudar esse critério,por exemplo, passando para ≤ 5, o modelo treinado com a definição antiga estará desalinhado e precisará ser refeito. A definição do target deve ser versionada e documentada. <br>  
**Target drift**: Mesmo sem mudar a definição, a proporção de Detratores pode variar ao longo do tempo, por uma crise logística, sazonalidade ou mudança de perfil de cliente. Um modelo treinado num período de baixo churn pode perder performance num período de crise. Por isso, o modelo precisa ser monitorado continuamente em produção.



*A variável nps_detrator será criada na fase de Data Preparation como,  
    nps_detrator = 1 se nps_score ≤ 6, caso contrário 0.*
